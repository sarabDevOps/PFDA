```python
#assignment03-pie
import pandas as pd
import matplotlib.pyplot as plt

# URL of the csv file
url = 'https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download'

#Step1: Load the data
data = pd.read_csv(url)

#Step2: Check input date has an 'Email' column
if 'Email' not in data.columns:
    raise ValueError("The CSV file must contain an 'Email' column.")

#Step3: Extract email domains using string.split
data['domain'] = data['Email'].str.split('@').str[1]

#Step4: Count occurrences of each domain
domain_counts = data['domain'].value_counts()

#Step5: Plot a pie chart
plt.figure(figsize=(10, 8))
plt.pie(domain_counts, labels=domain_counts.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 10})

#Step6: Enhance the chart aesthetics
plt.title('Distribution of Email Domains', fontsize=16, weight='bold')
plt.tight_layout()

#Step7: Display the pie chart
plt.show()
```
