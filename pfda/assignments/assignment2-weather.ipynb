```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

#Step1: Load the weather data from input file
file_path = r'D:\FB_friends\sarbajieeth_Ireland\weatherreadings1.csv'  
data = pd.read_csv(file_path)

#Step2: check the required column "dryBulbTemperature_Celsius" from input file 
if 'dryBulbTemperature_Celsius' not in data.columns or 'reportStartDateTime' not in data.columns:
    raise ValueError("The CSV file must contain 'reportStartDateTime' and 'dryBulbTemperature_Celsius' columns.")

#Step3: Convert reportStartDateTime to a datetime format
data['reportStartDateTime'] = pd.to_datetime(data['reportStartDateTime'])

#Step4: Sort data by reportStartDateTime
data = data.sort_values(by='reportStartDateTime')

#Step5: Plot the temperature over time
plt.figure(figsize=(12, 6))
plt.plot(data['reportStartDateTime'], data['dryBulbTemperature_Celsius'], color='blue', linewidth=1.5, label='Temperature (Celsius)')

#Step6: Enhance the plot aesthetics
plt.title('Temperature Over Time', fontsize=16, weight='bold')
plt.xlabel('Time', fontsize=12)
plt.ylabel('Temperature (Celsius)', fontsize=12)

#Step7: Format x-axis for datetime
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.xticks(rotation=45, fontsize=10)

plt.grid(visible=True, linestyle='--', alpha=0.6)
plt.legend(fontsize=10)
plt.tight_layout()

#Step8: Display the plot
plt.show()
```
